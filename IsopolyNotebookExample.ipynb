## 1) Isothermal Parker Solar Wind
First, we solve and plot isothermal Parker solar wind solutions where we set one free parameter such as $T_0 = T_p = T_e$, the coronal temperature, which is constant with radius out to infinity. This was the solution presented in the landmark [Parker 1958](https://ui.adsabs.harvard.edu/abs/1958ApJ...128..664P/abstract) paper.

### Code
The solution is retrieved with the inputs parameters :

```python
# Length of the output model
N = 1e4
L = 1.496e11      # set to 1au by default

# Polytropic indexes
gamma_p_max = 1
gamma_e_max = 1

# Coronal temperature
Tp0 = 1e6
Te0 = Tp0

# Isothermal radius (in solar radii)
r_iso_p = float('inf') 
r_iso_e = float('inf') 

# Expansion factor parameters
fm = 50
r_exp = 1.7          # in solar radii
sig_exp = 0.1       # in solar radii
```

where the inputs to change are : N is the number of points in the output python vector, Tp0[K] is the coronal temperature, fm is the final value of the expansion factor, r_exp[solar radii] and sig_exp[solar radii] are the expansion radius and length, respectively. The other inputs are set to an isothermal evolution. 

The outputs are the same as defined in the README file :

(r, n, u, Tp, Te, gamma_p, gamma_e, f, bol_super) with, in the respective order, radial distance, density, bulk speed, proton temperature, electron temperature, proton polytropic index, electron polytropic index, expansion factor,  and a bolean indicating the f-subsonic or f-supersonic type of solution.

### Plotting option
The plot options let the user choose which output quantity to display: f, gamma_p and gamma_e, a subplot with u, n, Tp and Te, and energy contributions. In this example, we plot everything except gamma_p and gamma_e, which are constant and equal to 1. 

```python
#########################################
# Plotting option 
plot_f = True
plot_gamma = False

plot_unT = True
plot_energy = True
#########################################
```

The different energetic contribution are computed based on the Bernoulli equation: 

\begin{align}
    \int_{r_1}^{r_2} m_p u \frac{d u}{d r} dr
    + \int_{r_1}^{r_2} \sum_{s = \{p,e\} } \bigg(  \frac{ 1 }{n} \: \frac{d P_s }{ d r} \bigg) dr 
    + \int_{r_1}^{r_2} m_p \frac{G M}{r^2} dr = cte . 
    \quad \Rightarrow \quad
    \underbrace{m_p \frac{u^2}{2}}_{E_c}
    + \underbrace{\int_{r_1}^{r_2} \sum_{s = \{p,e\} } \bigg(  \frac{ k_B}{n} \: \frac{d (n T_s) }{ d r} \bigg) dr}_{E_{th}} 
    - \underbrace{m_p \frac{G M}{r}}_{E_g} = cte .
\end{align}

The conservation of total energy attests that the solver calculates the momentum equation with respect to the physical consideration intrinsic to the plasma thermodynamics and evolution. The main script that runs the solver and takes all the inputs provides plots as follows :


```python
###############################################################
# Running the main function
(r, n, u, Tp, Te, gamma_p, gamma_e, f, bol_super) = ipsw.solve_isopoly(
                                        N, L, gamma_p_max, gamma_e_max, 
                                        Tp0, Te0, r_iso_p, r_iso_e,
                                        fm, r_exp, sig_exp, plot_f, 
                                        plot_gamma, plot_unT, 
                                        plot_energy)
###############################################################
```

### 3) Bi-fluid isopoly with single transition (f-subsonic)

In this example, the solar wind is assumed to be isothermal from $1r_\odot$ out to the isothermal radius $r_{iso}$, approximating temperature behavior in the corona. 

Beyond this radius the solar wind is assumed to be polytropic with index $\gamma_p$ for protons and $\gamma_e$ for electrons. 

As long as the solar wind goes supersonic inside the isothermal region and remains supersonic across the discontinuity in sound speed at the isothermal-polytropic boundary, then accelerating trans-sonic solution exists for $r \rightarrow \infty$.

```python
# Inputs of the model 
#########################################

# Length of the output model
N = 5e4
L = 1.496e11      # set to 1au by default

# Polytropic indexes
gamma_p_max = 1.45
gamma_e_max = 1.25

# Coronal temperature
Tp0 = 1.1e6
Te0 = 0.8e6

# Isothermal radius (in solar radii)
r_iso_p = 9 
r_iso_e = r_iso_p 

# Expansion factor parameters
fm = 30
r_exp = 1.9          # in solar radii
sig_exp = 0.1       # in solar radii
#########################################
# Plotting option 
plot_f = False
plot_gamma = False

plot_unT = True
plot_energy = False
#########################################

###############################################################
# Running the main function
(r, n, u, Tp, Te, gamma_p, gamma_e, f, bol_super) = ipsw.solve_isopoly(
                                        N, L, gamma_p_max, gamma_e_max, 
                                        Tp0, Te0, r_iso_p, r_iso_e,
                                        fm, r_exp, sig_exp, plot_f, 
                                        plot_gamma, plot_unT, 
                                        plot_energy)
###############################################################
```

### 3) Bi-fluid isopoly with double transition (f-supersonic)

This example takes the same thermal inputs parameters as for the privous example, but here we present the f-supersonic solution type that can be obtained modifying the super-radial expansion parameters ([Dakeyo et al. 2024b](https://arxiv.org/abs/2408.06155)). The work of pressure force is modified due to the deLaval nozzle effect, which leads for late and large expanding flux tube to potentially considerably enhance the acceleration in the Sun region. 

The parameters which influence the solution determination between f-subsonic and f-supersonic solutions are f_m, r_exp a,d sig_exp. A parametric study is presented in ([Dakeyo et al. 2024b](https://arxiv.org/abs/2408.06155)). 

It results that f-supersonic solution results generally in large final expansion (large f_m), late expansion radius (large r_exp) and small expansion length (small sig_exp). 
Changing only f_m from the previous example to larger value leads to:

```python
# Inputs of the model 
#########################################

# Length of the output model
N = 5e4
L = 1.496e11      # set to 1au by default

# Polytropic indexes
gamma_p_max = 1.45
gamma_e_max = 1.25

# Coronal temperature
Tp0 = 1.1e6
Te0 = 0.8e6

# Isothermal radius (in solar radii)
r_iso_p = 9 
r_iso_e = 15 

# Expansion factor parameters
fm = 70
r_exp = 1.9          # in solar radii
sig_exp = 0.1       # in solar radii
#########################################
# Plotting option 
plot_f = False
plot_gamma = False

plot_unT = True
plot_energy = False
#########################################

###############################################################
# Running the main function
(r, n, u, Tp, Te, gamma_p, gamma_e, f, bol_super) = ipsw.solve_isopoly(
                                        N, L, gamma_p_max, gamma_e_max, 
                                        Tp0, Te0, r_iso_p, r_iso_e,
                                        fm, r_exp, sig_exp, plot_f, 
                                        plot_gamma, plot_unT, 
                                        plot_energy)
###############################################################
```

## Inputs values recommandations
The values of the inputs affect the existence of a solar wind solution. In order to be as easy as possible in the case of existence of solution, we provide here values interval for which the numerical resolution can be operated nicely. The values are indicative and do not represent a strict interval for the existence of a solar wind solution. However, if all values are in the interval, increasing the number of points of the output vector up to $N = 10^5$ may help to find an existing solar wind solution.

$$
\left\{
    \begin{array}{c}
        N \geq 10^3 \quad \text{for} \quad L = 1au \\
        1 \: \leq \: (\gamma_p, \gamma_e) \: \leq \: 5/3 \\
         0.5 \: \text{MK} \: \lesssim \: (T_{p0}, T_{e0}) \: \lesssim \: 7 \: \text{MK} \\
         1.05 \times r_c \: \lesssim \: (r_{iso|p}, r_{iso|e}) \: \lesssim \: 25 \: r_\odot
    \end{array}
\right.
$$
